```python
import pandas as pd
import seaborn as sns

penguins = sns.load_dataset('penguins')

print(penguins)
```

```python
# penguins 데이터의 상위 10개 출력

penguins.head()
```

```python
# 부리 길이 55 이상 출력
penguins[penguins['bill_length_mm']>=55]
```

```python
# 연습 - 지느러미길이 180 미만 출력
penguins[penguins['flipper_length_mm']<180]
```

```python
# 연습 - 지느러미길이 200 이상 출력, 상위 3행만
penguins[penguins['flipper_length_mm']>=200].head(3)
```

```python
#자료 중 "성별: 암컷" 혹은 "서식지: Dream"에 해당하는 자료 추출
#합집합의 개념

penguins[penguins['sex']=='Female']
penguins[penguins['island']=='Dream']
```

```python
#자료 중 "성별: 암컷" 이면서 "서식지: Dream"에 해당하는 자료 추출
#교집합의 개념

conditions=(penguins['sex']=='Female') & (penguins['island']=='Dream')
penguins[conditions].head()
```

```python
#"가장 무거운 펭귄"을 추출하는 방법
penguins[penguins['body_mass_g'].max()==penguins['body_mass_g']]
```

```python
#연습 - 가장 가벼운 펭귄 추출.
#.min() 함수 사용
penguins[penguins['body_mass_g'].min()==penguins['body_mass_g']]
```

```python
# 무거운 펭귄 내림차순 정렬, 상위 10개 출력

penguins.sort_values(['body_mass_g'],ascending=False).head(10)
```

```python
#연습 - 가벼운 펭귄 내림차순 정렬, 상위 10개 출력

penguins.sort_values(['body_mass_g'],ascending=True).head(10)
```

```python
# 펭귄 몸무게 kg 단위로 출력
#상위 5개만 출력
penguins['body_mass_kg']=penguins['body_mass_g']/1000
penguins.head()
```

```python
# 연습 - 펭귄 지느러미 m 단위로 출력
# 상위 5개 출력
penguins['flipper_length_m']=penguins['flipper_length_mm']/1000
penguins.head()
```

```python
#종별로 몸무게 평균 출력
pd.DataFrame(penguins.groupby('species').mean()['body_mass_kg'])
```

```python
#연습 - 서식지별 지느러미 길이 평균 출력
pd.DataFrame(penguins.groupby('island').mean()['flipper_length_mm'])
```

```python
# 펭귄 데이터세트 전체에서 species 컬럼의 Adelie와 Gentoo만 필터링 후 출력
# 특정 조건 선택 시 isin함수를 사용
conditions=penguins['species'].isin(['Adelie','Gentoo'])
penguins[conditions]
```

```python
# 연습 - 펭귄 데이터세트 전체에서 서식지 Deream만 필터링 후 출력
# 특정 조건 선택 시 isin함수를 사용
# 상위 5개 컬럼만 출력
conditions=penguins['island'].isin(['Dream'])
penguins[conditions].head()
```

```python
#전체 펭귄 데이터의 몸무게 평균 출력
mean_mass=penguins['body_mass_kg'].mean()
mean_mass
```

```python
#연습 - 전체 펭귄 데이터의 지느러미 길이 평균 출력
mean_flipper=penguins['flipper_length_m'].mean()
mean_flipper
```

```python
# 몸무게가 전체 펭귄 몸무게 평균보다 많은 데이터만 필터링.
penguins[penguins['body_mass_kg']>=penguins['body_mass_kg'].mean()]
```

```python
# 연습 - 지느러미 길이가 전체 펭귄 지느러미 길이 평균보다 많은 데이터만 필터링.
# 상위 5개 출력
penguins[penguins['flipper_length_mm']>=penguins['flipper_length_mm'].mean()].head()
```

```python
#선택적 컬럼 출력 - species, island, bodymass (kg) 만 출력
penguins[['species','island','body_mass_kg']]
```

```python
# 연습 - 선택적 컬럼 출력 - bill_depth_mm,	flipper_length_mm,	body_mass_g만 출력

penguins[['bill_depth_mm','flipper_length_mm','body_mass_g']].head()
```

```python
# 몸무게가 적은 순서대로 10개의 행 출력
# species, island, bodymass (kg) 만 출력

penguins[['species','island','body_mass_kg']].sort_values('body_mass_kg',ascending=True).head(10)
```

결측데이터 처리

```python
#결측데이터 확인- body_mass_kg 컬럼

penguins[penguins['body_mass_kg'].isna()]
```

```python
# 컬럼별 결측데이터 개수 확인
penguins.isnull().sum()
```

```python
#결측데이터 확인- body_mass_kg 컬럼

penguins[penguins['sex'].isna()]
```

```python
#drop 함수 이용
#Sex 컬럼 전체 삭제

penguins.drop('sex',axis=1,inplace=True)
```

```python
#컬럼 이름 확인
penguins.columns
```

```python
#삭제된 데이터 프레임 확인

penguins.head()
```

```python
# mass(kg) 의 최빈값 찾은 후 결측데이터에 대체
penguins['body_mass_kg'].mode()
```

```python
# mass(kg) 컬럼의 결측데이터를 인덱싱 0 값 3.8로 대체
# 결측데이터가 존재했던 3번 확인

penguins['body_mass_kg'].fillna(penguins['body_mass_kg'].mode()[0],inplace=True)
penguins.head()
```

```python
#body_mass_kg 결측데이터 행 확인
penguins[penguins['body_mass_kg'].isna()]
```

```python
# 결측데이터가 존재했던 339번 확인
penguins.loc[339, 'body_mass_kg']
```

```python
# body_mass_kg 결측데이터가 대체된 데이터 프레임의 전체 결측데이터 확인

penguins.isnull().sum()
```

```python
# 연습 - bill_length_mm 결측데이터 행 확인
penguins[penguins['bill_length_mm'].isna()]
```

```python
# 연습 - bill_length_mm 의 최빈값 찾은 후 결측데이터에 대체
penguins['bill_length_mm'].mode()
```

```python
# 연습 - bill_length_mm 컬럼의 결측데이터를 최빈값으로 대체
penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mode()[0],inplace=True)
```

```python
# 연습 - 결측데이터 행 확인
penguins[penguins['bill_length_mm'].isna()]
```

```python
# 연습 - bill_length_mm 결측데이터가 대체된 데이터 프레임의 전체 결측데이터 확인

penguins.isnull().sum()
```
